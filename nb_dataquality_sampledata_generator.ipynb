Code to create the sample data file

```python
import pandas as pd
import numpy as np

# Function to generate sample dataset
def generate_sample_dataset(num_rows):
    np.random.seed(42)  # For reproducibility
    data = {
        "CustomerID": range(1, num_rows + 1),
        "Name": [f"Name_{i}" if i % 10 != 0 else None for i in range(1, num_rows + 1)],  # Some missing values
        "Email": [
            f"user{i}@example.com" if i % 15 != 0 else f"user{i}@"
            for i in range(1, num_rows + 1)
        ],  # Some invalid emails
        "Age": [np.random.randint(18, 70) if i % 20 != 0 else None for i in range(1, num_rows + 1)],  # Missing values
        "SignupDate": [
            f"2023-{np.random.randint(1, 13):02d}-{np.random.randint(1, 29):02d}" if i % 25 != 0 else "InvalidDate"
            for i in range(1, num_rows + 1)
        ],  # Some invalid dates
        "IsActive": [
            np.random.choice([True, False, "Yes", None], p=[0.4, 0.4, 0.1, 0.1]) for _ in range(num_rows)
        ],  # Mixed boolean representations and missing values
    }
    return pd.DataFrame(data)

# Generate dataset with 1000 rows
num_rows = 1000
dataset = generate_sample_dataset(num_rows)

# Save dataset to your personal workspace Lakehouse
output_path = "abfss://Rachit_Personal@onelake.dfs.fabric.microsoft.com/Rachit_Personal_Lakehouse.Lakehouse/Files/sample_data_quality_test_large.csv"
dataset.to_csv(output_path, index=False)

print(f"Dataset with {num_rows} rows created and saved to {output_path}")
```
